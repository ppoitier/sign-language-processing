```python
import numpy as np
import pandas as pd

from sign_language_tools.player.video_player import VideoPlayer
from sign_language_tools.pose.mediapipe.edges import POSE_EDGES, HAND_EDGES
```

```python
root = "Z:/data/lsa64"
samples = pd.read_csv(f"{root}/index.csv")
samples
```

```python
def display_sample(sample_id: str, speed=1.0):
    sample_poses = {region: np.load(f"{root}/poses/{region}/{sample_id}.npy") for region in ['pose', 'left_hand', 'right_hand']}
    player = VideoPlayer()
    player.attach_video(f"{root}/videos/{sample_id}.mp4", name='video')
    player.attach_poses(sample_poses['pose'], POSE_EDGES, parent_name='video')
    player.attach_poses(sample_poses['left_hand'], HAND_EDGES, parent_name='video')
    player.attach_poses(sample_poses['right_hand'], HAND_EDGES, parent_name='video')
    player.play(speed=speed)

# display_sample('001_001_001')
# display_sample('001_001_002')
# display_sample('001_001_003')
# display_sample('001_001_004')
# display_sample('001_001_005')
```

```python
def display_samples_from_class(class_nb: int, n=5):
    sample_ids = samples.loc[samples['class'] == class_nb, 'id'].sample(n=n).to_list()
    for sample_id in sample_ids:
        display_sample(sample_id, speed=2.0)


def display_samples_from_signer(signer_id: int, n=5):
    sample_ids = samples.loc[samples['signer_id'] == signer_id, 'id'].sample(n=n).to_list()
    for sample_id in sample_ids:
        display_sample(sample_id, speed=2.0)

for i in range(1, 11):
    display_samples_from_signer(i)
```
